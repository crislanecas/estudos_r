# Listando pacotes

```r
# Mostra uma lista com os pacotes já instalados.
installed.packages()
```

# Instalando pacotes

```r
# Instala o pacote especificado como parâmetro.
install.packages("tidyverse")
```

# Carregando pacotes

```r
# Carrega o pacote especificado como parâmetro.
library(tidyverse)
```

# Atualizando pacotes

```r
# Atualiza do pacote tidyverse.
tidyverse_update()
```

```r
# Atualiza os pacotes especificados como argumentos.
update.packages(c("ggplot2", "ragg", "rvest"))
```

```r
# Atualiza todos os pacotes.
update.packages()
```
