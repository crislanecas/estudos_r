# Função as_tibble()
Tibble é semelhante a um data frame padrão, porém projetada para facilitar a manipulação e visualização de dados, possui algumas melhorias e diferenças importantes.  

**Impressão melhorada:** Quando você imprime uma tibble, apenas as primeiras 10 linhas e todas as colunas que cabem na largura da tela são exibidas. Isso evita a exibição excessiva de dados e facilita a visualização de grandes conjuntos de dados.

**Subconjuntos consistentes:** A indexação de tibbles retorna sempre uma tibble, mesmo que resulte em apenas uma linha ou uma coluna. Isso torna o comportamento mais consistente em comparação com os data frames padrão do R.

**Nomes de colunas:** Os nomes das colunas são tratados de forma mais rigorosa, evitando a conversão automática de nomes de colunas em nomes de variáveis inválidos.

**Tipos de dados:** A tibble é mais rigorosa em relação aos tipos de dados, evitando a conversão automática de strings em fatores e outros comportamentos indesejados.

**Eficiência:** Embora a tibble seja mais eficiente em termos de desempenho e uso de memória em comparação com os data frames padrão, a diferença pode não ser significativa para conjuntos de dados pequenos a moderados.

```r
# Cria uma tibble.
as_tibble(diamonds)
```

# Função rename()

```r
# Renomeia a coluna especificada no dataset.
rename(diamonds, carat_new = carat)
```

```r
# Renomeia variárias colunas especificadas no dataset.
rename(diamonds, carat_new = carat, cut_new = cut)
```

# Função summarize()

```r
# Cria um resumo estatístico do dataset, conforme especificado.
summarize(diamonds, mean_carat = mean(carat))
```

# Função ggplot()

```r
# Cria um gráfico de disperção, conforme argumentos informados.
ggplot(data = diamonds, aes(x = carat, y = price)) +
  geom_point()
```

```r
# Cria um gráfico de disperção, incluindo o parâmetro cor para diferenciar os cortes.
ggplot(data = diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point()
```

```r
# Cria um gráfico de disperção, separando para cada tipo de corte.
ggplot(data = diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point() +
  facet_wrap(~cut)
```
